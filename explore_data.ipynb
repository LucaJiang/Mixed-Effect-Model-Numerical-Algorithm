```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Read in the data
data = pd.read_table('data/XYZ_MoM.txt', sep='\t', header=0).values
y = data[:, 0].reshape(-1, 1)
Z = data[:, 1:31]
X = data[:, 31:]
```

```python
# Plot the data
plt.figure(figsize=(10, 10))
plt.subplot(3, 1, 1)
# density plot of y
sns.distplot(y, hist=False, rug=True)
# text mean and variance of y
plt.text(y.mean(), 0.1, 'mean: {:.2f}'.format(y.mean()))
plt.text(y.mean(), 0.08, 'var: {:.2f}'.format(y.var()))
plt.axvline(y.mean(), color='r', linestyle='--')
plt.xlabel('y')
plt.ylabel('density')

plt.subplot(3, 1, 2)
# density plot of column of X
X_col_mean = X.mean(axis=0)
X_col_var = X.var(axis=0)
sns.distplot(X_col_mean, hist=False, rug=True, label='mean')
sns.distplot(X_col_var, hist=False, rug=True, label='var')
# text mean and variance of X
plt.text(X_col_mean.mean(), 3e4, 'mean: {:.2f}'.format(X_col_mean.mean()))
plt.text(X_col_mean.mean(), 2e4, 'var: {:.2f}'.format(X_col_var.mean()))
plt.xlabel('X')
plt.ylabel('density')
plt.legend()

plt.subplot(3, 1, 3)
# density plot of column of Z
Z_col_mean = Z.mean(axis=0)
Z_col_var = Z.var(axis=0)
sns.distplot(Z_col_mean, hist=False, rug=True, label='mean')
sns.distplot(Z_col_var, hist=False, rug=True, label='var')
# text mean and variance of Z
plt.text(0.4, 8, 'mean: {:.2f}'.format(Z_col_mean.mean()))
plt.text(0.4, 6, 'var: {:.2f}'.format(Z_col_var.mean()))
plt.xlabel('Z')
plt.ylabel('density')
plt.legend()

plt.suptitle('Distribution of y, X, Z')
plt.tight_layout()
plt.savefig('img/explore_data.png')
plt.show()
```
